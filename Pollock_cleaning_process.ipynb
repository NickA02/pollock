## This is a notebook to explore cleaning methods for the arXiv dataset

```python
#import libraries 
import numpy as np
import pandas as pd
import json
import os.path
```

```python
# Get a list of dicts and convert into a pandas df.
arXiv_data = []

# Replace file path with where you have metadata
with open('/Users/edwardbaker/Desktop/NLP/arxiv-metadata-oai-snapshot.json', 'r') as file:
    # grab the first 10 metadata
    for i in range(10):
        line = next(file).strip()
        arXiv_data.append(json.loads(line))
    
df = pd.DataFrame.from_records(arXiv_data) #convert data to dataframe
```

```python
#select required columns 
df = df[["id", "abstract"]]
```

### We probably want to remove the author, title, etc. as well... which is also in metadata* 

```python
# make the ids indexible
df = df.set_index('id')
df.head()
```

```python
text = {}

with open('/Users/edwardbaker/Desktop/NLP/0704.0003v2.txt', 'r', encoding='cp1252') as file:
    path = os.path.basename('/Users/edwardbaker/Desktop/NLP/0704.0003v2.txt')
    dictkey = path[0:9]
    text[dictkey] = " ".join(file.readlines())
    
textframe = pd.DataFrame.from_dict(text, orient='index')
```

```python
textframe = textframe.rename({0: 'text'},axis='columns')
textframe.head()
```

### Do other processing before removing abstract, this will help w/ match

```python
for row in textframe.index:
    # abstract and text corresponding to the id
    abstract = df.loc[row]["abstract"]
    text = textframe.loc[row][text]
    
    # remove the abstract, use fuzzy match
    text = text.replace(abstract, "")
    # replace in df
#     textframe.loc[row][0] = text
    
    
```
